```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import pickle
```

```python
df_imoveis = pd.read_csv('preco_casas.csv')
```

```python
df_imoveis.head()
```

```python
df_imoveis.info()
```

# Modelo

```python
X = df_imoveis.drop('preco', axis = 1).copy()
y = df_imoveis.preco.copy()
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
reg = LinearRegression()
reg.fit(X_scaled,y)
```

```python
casa_nova = [[100, 3, 3, 2]]
casa_nova = scaler.transform(casa_nova)
reg.predict(casa_nova)
```

```python
# save the scaler
pickle.dump(scaler, open('scaler.pkl', 'wb'))
```

```python
filename = 'modelo_treinado.sav'
pickle.dump(reg, open(filename, 'wb'))
```
